### Selecting rows from a DataFrame

- **iloc** - searches using index positions
- **loc** - searches using index labels 

loc is typically used for label indexing and can access multiple columns, while . iloc is used for integer indexing

```python
# creating a sample dataframe
data = pd.DataFrame({'Brand': ['Maruti', 'Hyundai', 'Tata',
                               'Mahindra', 'Maruti', 'Hyundai',
                               'Renault', 'Tata', 'Maruti'],
                     'Year': [2012, 2014, 2011, 2015, 2012,
                              2016, 2014, 2018, 2019],
                     'Kms Driven': [50000, 30000, 60000,
                                    25000, 10000, 46000,
                                    31000, 15000, 12000],
                     'City': ['Gurgaon', 'Delhi', 'Mumbai',
                              'Delhi', 'Mumbai', 'Delhi',
                              'Mumbai', 'Chennai',  'Ghaziabad'],
                     'Mileage':  [28, 27, 25, 26, 28,
                                  29, 24, 21, 24]})
 
# displaying the DataFrame
display(data)
```

```python
# selecting cars with brand 'Maruti' and Mileage > 25
display(data.loc[(data.Brand == 'Maruti') & (data.Mileage > 25)])
```

```python
# selecting range of rows from 2 to 5
display(data.loc[2: 5])
```

```python
# updating values of Mileage if Year < 2015
data.loc[(data.Year < 2015), ['Mileage']] = 22
display(data)
```

```python
# selecting 0th, 2nd, 4th, and 7th index rows
display(data.iloc[[0, 2, 4, 7]])
```

```python
# selecting rows from 1 to 4 and columns from 2 to 4
display(data.iloc[1: 5, 2: 5])
```

```python
# find which player has won most potm -> in finals and qualifiers

import pandas as pd

# Assuming your dataset is stored in a DataFrame called 'df'
# Replace 'df' with your actual DataFrame name if it's different

# Filtering for finals and qualifiers
finals_and_qualifiers = ipl[ipl['Season'].str.contains('Final|Qualifier')]

# Counting the occurrences of each player winning POTM
player_potm_count = finals_and_qualifiers['Player_of_Match'].value_counts()

# Finding the player with the most POTM awards
#most_potm_player = player_potm_count.idxmax()
most_potm_count = player_potm_count.max()

print(f"The player with the most POTM awards in finals and qualifiers is {player_potm_count} with {most_potm_count} awards.")
```

```python
# Toss decision plot

import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataset is stored in a DataFrame called 'df'
# Replace 'df' with your actual DataFrame name if it's different

# Counting the toss decisions
toss_decisions = ipl['TossDecision'].value_counts()

# Creating a bar plot
plt.figure(figsize=(8, 6))
toss_decisions.plot(kind='bar', color=['blue', 'green'])
plt.title('Toss Decisions by Teams')
plt.xlabel('Toss Decision')
plt.ylabel('Number of Matches')
plt.xticks(rotation=0)
plt.legend(['Fielding', 'Batting'])
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd

# Assuming your dataset is stored in a DataFrame called 'df'
# Replace 'df' with your actual DataFrame name if it's different

# Grouping by 'Team1' and counting matches
team1_matches = ipl.groupby('Team1').size().reset_index(name='MatchesPlayed')

# Grouping by 'Team2' and counting matches
team2_matches = ipl.groupby('Team2').size().reset_index(name='MatchesPlayed')

# Combining the results for both columns
total_matches = pd.concat([team1_matches, team2_matches], ignore_index=True)
total_matches = total_matches.groupby('Team1')['MatchesPlayed'].sum().reset_index()

# Renaming the 'Team1' column to 'Team' for clarity
total_matches.rename(columns={'Team1': 'Team'}, inplace=True)

print(total_matches)
```
